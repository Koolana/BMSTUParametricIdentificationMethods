# Задание 1. Дискретное преобразование фурье

```python
import numpy as np
import math 
import matplotlib.pyplot as plt
```

## Входная функция вида: $$y = e^x + x^3$$

```python
x = np.arange(0, 3, 0.01)
y = [math.e ** i + i ** 3 for i in x]
```

## Добавление нормальных шумов

```python
y_err = y + np.random.normal(scale=2, size=len(y))
```

## Вывод функции
- зеленая линия - сама функция
- синяя линия - зашумленная функция

```python
plt.figure(figsize=(20,10))
plt.plot(x,y,"g")
plt.plot(x,y_err,"b")
plt.title('Step Responsse ') 
plt.ylabel('Amplitude')
plt.xlabel('Time(sec)')
plt.grid(True)
plt.show()
```

## Преобразование Фурье от зашумленной функции и отсечение лишних частот

```python
frequency = np.fft.fft(y_err)

threshold = 5

frequency_clean = frequency.copy()
frequency_clean[threshold:len(frequency)-5] = 0
```

## Вывод спектров 
- зеленая линия - первоначальный спектр после преобразования Фурье
- синяя линия - спектр после отсечения

```python
plt.figure(figsize=(20,10))
plt.plot(np.arange(0, len(frequency), 1),frequency,"g")
plt.plot(np.arange(0, len(frequency_clean), 1),frequency_clean,"b")
plt.grid(True)
plt.show()
```

## Обратное преобразование Фурье усеченного спектра

```python
y_new = np.fft.ifft(frequency_clean)
```

## Вывод функции после обратного преобразования Фурье
- зеленая линия - первоначальная функция
- синяя линия - зашумленная функция
- красная линия - восстановленая функция после обратного преобразования Фурье

```python
plt.figure(figsize=(20,10))
plt.plot(x,y,"g")
plt.plot(x,y_err,"b")
plt.plot(x,y_new,"r")
plt.title('Step Responsse ') 
plt.ylabel('Amplitude')
plt.xlabel('Time(sec)')
plt.grid(True)
plt.show()
```
